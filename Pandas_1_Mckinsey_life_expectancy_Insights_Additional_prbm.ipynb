#Q4. Employee records

Given a dataframe consisting of Name, employment_type, department, and p_id for employees.

Complete the function del_dupl() that

Deletes the records of a given name, where name is a string.
Deletes the duplicate p_id rows and keeps only the last entry as it contains the latest employment details of the candidate.
Deletes the column p_id as the organization is moving to another fintech app, and returns this dataframe.
There are no two people with the same p_id in the organization.
If there are records with the same p_id then it should be assumed that the person has updated their profile (designation) in the organization.

```python
def del_dupl(df,name):
    ''' df is a dataframe with columns ['Name', 'employment_type', 'department', 'p_id']
        name is a string.
        Output a dataframe is expected to be returned'''

    # YOUR CODE GOES HERE

    # delete the record with name
    ind = df[df['Name']==name].index
    df.drop(ind, inplace = True)

    # drop the column 'p_id'
    df.drop(['p_id'],axis=1,inplace=True)
    # drop the duplicates, preserving the last one with same name

    df.drop_duplicates(subset=['Name'],keep='last',inplace=True)
    return df
```

Given a data frame as shown below, return the series of male names in the age group of 23-30 (inclusive of both 23 and 30).

```python
import pandas as pd
def specific_customers(df):
    ''' df is a dataframe with columns ['name', 'profession', 'gender', 'age', 'review', 'rating']
        Output -> A series with required names is expected to be returned'''
    # YOUR CODE GOES HERE

    # Select male professionals
    df_1 = df[df['gender']=='male']

    # Filter for age group 23-30
    df_new = df_1[df_1['age'].between(23,30)]

    return df_new['name']
```
