#7. Categorise ages


1. If age<18: kid
2. age >=18 and <=50: adult
3. age>50 senior

```python
data = {'name':["Ram","Shyam","Mukesh","Suresh"],'age':[10,18,60,50]}
```

```python
df5 = pd.DataFrame(data)
```

```python
df5
```

```python
def age_cat(x):
  if x < 18:
    return 'kid'
  elif x>= 18 and x <= 50:
    return 'adult'
  elif x> 50:
    return 'senior'
```

```python
df5['age'] = df5['age'].apply(age_cat)
```
